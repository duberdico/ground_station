```python
from sklearn.datasets import load_iris
import seaborn as sns
import random
import pandas as pd
import numpy as np
import subprocess
sns.set(style="ticks", color_codes=True)
import MIDAS
```

```python
n = 1000
sigma = 0.3
p_missing = 0.1
```

```python
iris = sns.load_dataset("iris")
print(iris.head())
```

```python

temp = pd.DataFrame(data=None, columns=iris.columns)
for i in range(n):
    j = random.randrange(len(iris))
    cur = iris.iloc[j].copy()
    cur['sepal_length'] = random.gauss(cur['sepal_length'],sigma)
    cur['sepal_width'] = random.gauss(cur['sepal_width'],sigma)
    cur['petal_length'] = random.gauss(cur['petal_length'],sigma)
    temp = temp.append(cur)
iris = iris.append(temp,ignore_index=True)
```

```python
 g = sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
```

```python
I = iris.copy()
m = random.sample(range(len(iris)),int(p_missing*len(iris)))
n = random.choice(range(len(iris.columns)))
I.loc[m,iris.columns[n]] = np.NaN
```
